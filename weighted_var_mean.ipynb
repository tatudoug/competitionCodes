```python
import numpy as np
import random
```

```python
vec1 = []
for num in np.arange(1000):
    vec1.append ( random.gauss(0, 1))
    
for num in np.arange(1000):
    vec1.append ( random.gauss(0, 2))
    
for num in np.arange(1000):
    vec1.append ( random.gauss(1, 1))
    
for num in np.arange(1000):
    vec1.append ( random.gauss(0, 1))
    
for num in np.arange(1000):
    vec1.append ( random.gauss(1, 2))

vec1 = np.array(vec1)
```

```python
mean = []
variance = []

mean.append(random.gauss(0, 1))
variance.append( random.gauss(0, 1))

alpha = 0.05

for x in np.arange(len(vec1)):
    diff = vec1[x] - mean[x]
    incr = alpha * diff
    mean.append( mean[x] + incr)
    variance.append(  (1-alpha)*(variance[x]+diff*incr) )
    
```

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.arange(len(mean))
x2 = np.arange(len(variance))

y1 = mean
y2 = variance

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('mean')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('variance')

plt.show()
```
